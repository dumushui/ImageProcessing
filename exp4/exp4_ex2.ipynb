2.利用适当方法对1.输出的图像进行增强（例如直方图均衡化，线性变换等）；

```python
from skimage import io
import numpy as np
import matplotlib.pyplot as plt
gaussinan = io.imread(r'.\pic\gaussiannoiseimage.bmp',0)
saltpepoer = io.imread(r'.\pic\saltpepoernoiseimage.bmp')
from skimage.morphology import disk
import skimage.filters.rank as sfr
def mean_img(img,n):
    imgt = sfr.mean(img, disk(n))
    return imgt
def median_img(img,n):
    imgt = sfr.median(img, disk(n))
    return imgt
saltpepoeraftermedian = median_img(saltpepoer,2)
gaussinanaftermean = mean_img(gaussinan,2)
plt.subplot(121);plt.title('saltpepoeraftermedian');plt.imshow(saltpepoeraftermedian, cmap ='gray')
plt.axis("off")
plt.subplot(122);plt.title('gaussinanaftermean');plt.imshow(gaussinanaftermean, cmap ='gray')
plt.axis("off")
```

```python
sarr = saltpepoeraftermedian.flatten()
plt.figure("hist")
n, bins, patches = plt.hist(sarr, bins=256)  
plt.show() 
```

上图为saltpepoeraftermedian的直方图

```python
garr = gaussinanaftermean.flatten()
plt.figure("hist")
n, bins, patches = plt.hist(garr, bins=256)  
plt.show() 
```

上图为gaussinanaftermean的直方图,接下来对这副图像作线性变换以增强图像

```python
from skimage import exposure
gaussinanaftermean1=exposure.equalize_hist(gaussinanaftermean)
#io.imshow(gaussinanaftermean1)
plt.subplot(121);plt.title('gaussinanaftermean');plt.imshow(gaussinanaftermean, cmap ='gray')
plt.axis("off")
plt.subplot(122);plt.title('gaussinanaftermean1');plt.imshow(gaussinanaftermean1, cmap ='gray')
plt.axis("off")
```

增强之后的图像如上图，它的直方图如下图

```python
garr2 = gaussinanaftermean1.flatten()
plt.figure("hist")
n, bins, patches = plt.hist(garr2, bins=256)  
plt.show() 
```

```python
saltpepoeraftermedian2 = exposure.equalize_adapthist(saltpepoeraftermedian, clip_limit=0.03)
#io.imshow(saltpepoeraftermedian2)
plt.subplot(121);plt.title('saltpepoeraftermedian');plt.imshow(saltpepoeraftermedian, cmap ='gray')
plt.axis("off")
plt.subplot(122);plt.title('saltpepoeraftermedian2');plt.imshow(saltpepoeraftermedian2, cmap ='gray')
plt.axis("off")
```

对椒盐噪声图像做直方图均衡化的图像增强，上图即为结果，下图为直方图

```python
sarr2 = saltpepoeraftermedian2.flatten()
plt.figure("hist")
n, bins, patches = plt.hist(sarr2, bins=256)  
plt.show() 
```
