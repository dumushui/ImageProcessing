2.选择一幅RGB彩色图像（图像RGBimageA），分别显示出原图像和R、G、B三个分量图像（用subplot函数显示在同一窗口中），观察对比它们的特点，体会不同颜色所对应的R、G、B分量的不同之处。

```python
import numpy as np
import cv2
import matplotlib.pyplot as plt
```

- 因为skimage中并没有能够实现任务的函数，现在换用opencv来实现图像分离
- 因为是在jupyter notebook中写的，cv2.show()函数没办法使用，使用了matplotlib来代替

```python
img = cv2.imread(r'.\image\luna.jpg')
#cv2.imshow('image',img)
#cv2.waitKey(0)
plt.imshow(img)
plt.axis("off")#去除坐标轴
plt.show()
```

- 因为opencv默认的imread是以BGR的方式进行存储的,而matplotlib的imshow默认则是以RGB格式展示,所以此处我们必须对图片的通道进行转换
- 转换方法是使用cvtColor这个函数，将BGR通道转化为RGB,因此就是cv2.COLOR_BGR2RGB

```python
img2 = cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
plt.imshow(img)
plt.axis("off")#去除坐标轴
plt.show()
```

- 另外一种正确显示的方法是使用imutils包，这个包提供一系列便捷功能进行基本的图像处理功能，如平移，旋转，缩放，骨架，matplotlib图像显示，排序的轮廓，边缘检测，让你更容易使用OpenCV和Python
- 这样做的好处是保留原来的BGR通道图像，此处使用的是imutils.opencv2matplotlib()

```python
import imutils
plt.imshow(imutils.opencv2matplotlib(img))
plt.axis("off")#去除坐标轴
plt.show()
```

```python
img = cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
(B,G,R) = cv2.split(img)
plt.subplot(221); plt.imshow(img)
plt.axis("off")
plt.subplot(222); plt.imshow(R, cmap ='gray')
plt.axis("off")
plt.subplot(223); plt.imshow(G, cmap ='gray')
plt.axis("off")
plt.subplot(224); plt.imshow(B, cmap ='gray')
plt.axis("off")
plt.show()
```
