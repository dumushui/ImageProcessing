3、给图像C加入高斯噪声，分别采用不同大小的模板对加有噪声的图像进行均值滤波，用一个图像处理对话框（subplot）显示原图像、加有噪声的图像及均值滤波的图像。比较结果。

```python
from skimage import io
import matplotlib.pyplot as plt
img = io.imread('.\pic\c.tif')
io.imshow(img)
```

```python
from skimage import util
img2=util.random_noise(img,mode='gaussian')
io.imshow(img2)
```

使用均值滤波器（mean) : 返回图像局部区域内的均值，用此均值取代该区域内所有像素值。

```python
from skimage.morphology import disk
import skimage.filters.rank as sfr
def mean_img(img,n):
    imgt = sfr.mean(img2, disk(n))
    return imgt
```

```python
import matplotlib.pyplot as plt
plt.subplot(221);plt.title('img');plt.imshow(img, cmap ='gray')
plt.axis("off")
plt.subplot(222);plt.title('img_noise');plt.imshow(img2, cmap ='gray')
plt.axis("off")
plt.subplot(223);plt.title('r = 2');plt.imshow(mean_img(img,2), cmap ='gray')
plt.axis("off")
plt.subplot(224);plt.title('r = 5');plt.imshow(mean_img(img,5), cmap ='gray')
plt.axis("off")
plt.show()
```
