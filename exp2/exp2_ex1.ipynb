* 改变灰度值调整范围，观察调整后的图像的变化及其与原图像的差别，调整后的灰度直方图的变化及其与原灰度直方图的区别；

```python
from skimage import io
import matplotlib.pyplot as plt
img = io.imread('.\pic\pout.tif')
io.imshow(img)
```

显示图像A及其灰度直方图

```python
arr = img.flatten()
plt.figure("hist")
n, bins, patches = plt.hist(arr, bins=256)  
plt.show() 
```

显示图像A及其灰度直方图。将它的灰度值调整到[0,1]之间，调整后的灰度直方图与原灰度直方图的区别；

```python
plt.figure("hist")
n, bins, patches = plt.hist(arr, bins=256, normed=1 )  
plt.show() 
```
