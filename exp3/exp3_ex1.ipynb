1、	读取图像A(lena.tiff)和B(rice.tif)，显示这两幅图像，对图像作傅立叶变换，显示图像的傅里叶幅度谱和相位谱。做傅立叶逆变换，显示重建图像，看是否与原图像相同。

```python
from skimage import io
import numpy as np
import matplotlib.pyplot as plt
lena = io.imread(r'.\pic\lena.tiff',0)
rice = io.imread(r'.\pic\rice.tif')
plt.subplot(121);plt.title('lena');plt.imshow(lena, cmap ='gray')
plt.axis("off")
plt.subplot(122);plt.title('rice');plt.imshow(rice, cmap ='gray')
plt.axis("off")
```

使用np.fft.fftshift（）生成lena位移后的fft，并取绝对数的对数来让转换后的矩阵可以显示

```python
lena_ft = np.fft.fft(lena)
fshift = np.fft.fftshift(lena_ft)
lena_ft = np.log(np.abs(lena_ft))
fshift = np.log(np.abs(fshift))
plt.subplot(121),plt.imshow(lena,'gray'),plt.title('lena')
plt.subplot(122),plt.imshow(fshift,'gray'),plt.title('shiftft')
```

使用np.fft.fftshift（）生成lena位移后的fft，上图只显示了频域的幅度信息，这里用np.angle取得图像的相位信息

```python
lena_ft = np.fft.fft(lena)
fshift = np.fft.fftshift(lena_ft)
ph_f = np.angle(lena_ft)
ph_fshift = np.angle(fshift)
plt.subplot(121),plt.imshow(lena,'gray'),plt.title('lena')
plt.subplot(122),plt.imshow(ph_fshift,'gray'),plt.title('center-angle')
```
