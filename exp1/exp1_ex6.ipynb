6、自行下载两幅的图像，进行相加、相减、相乘的代数运算（注意运算时候图像的尺寸、类型等，如果不同，应使将尺寸较大的图像裁剪或缩放），显示运算之前和运算之后的图像。

```python
from skimage import io
from skimage import transform
import matplotlib.pyplot as plt
img = io.imread(r'.\image\luna.jpg')
img2 = io.imread(r'.\image\view.jpg')
plt.subplot(121);plt.title('img');plt.imshow(img)
plt.axis("off")
plt.subplot(122);plt.title('img2');plt.imshow(img2)
plt.axis("off")
```

- 因为两张图片的尺寸不一，故需要修改尺寸，此处使用的skimage.transform.resize()修改的

```python
temp = img2
img2 = transform.resize(img2, (500, 500))
plt.subplot(121);plt.title('img2');plt.imshow(temp)
plt.axis("off")
plt.subplot(122);plt.title('resized');plt.imshow(img2)
plt.axis("off")
```

```python
plus_result = img + img2
substract_result = img - img2 
times_result = img * img2
plt.subplot(221);plt.title('img');plt.imshow(img)
plt.axis("off")
plt.subplot(222);plt.title('plus');plt.imshow(plus_result)
plt.axis("off")
plt.subplot(223);plt.title('substract');plt.imshow(substract_result)
plt.axis("off")
plt.subplot(224);plt.title('times');plt.imshow(times_result)
plt.axis("off")
plt.show()
```
