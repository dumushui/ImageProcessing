5.给灰度图像加入高斯噪声，并显示。
- 运用 for 循环，分别将 5 幅、50 幅和 500 幅加有随机高斯噪声的图像进行相加并求其平均值。
- 将几种求平均后的图像显示在同一图像对话框中，比较其结果

```python
from skimage import util
from skimage import io
img = io.imread(r'.\image\luna.jpg',-1)
img2 = util.random_noise(img,mode='gaussian')
io.imshow(img2)
```

```python
def average_img(img,n):
    imgt = util.random_noise(img,mode='gaussian')
    for x in range(n-1):
        imgt += util.random_noise(img,mode='gaussian')
    imgt = imgt/n
    return imgt
```

- 定义了一个循环产生并叠加最后平均的函数，参数img是原始图片，n是循环次数

```python
io.imshow(average_img(img,5))
```

```python
import matplotlib.pyplot as plt
plt.subplot(221);plt.title('img');plt.imshow(img, cmap ='gray')
plt.axis("off")
plt.subplot(222);plt.title('5');plt.imshow(average_img(img,5), cmap ='gray')
plt.axis("off")
plt.subplot(223);plt.title('10');plt.imshow(average_img(img,10), cmap ='gray')
plt.axis("off")
plt.subplot(224);plt.title('100');plt.imshow(average_img(img,100), cmap ='gray')
plt.axis("off")
plt.show()
```
